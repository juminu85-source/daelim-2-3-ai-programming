```python
# 할인율 계산
# discount: 0.0~1.0 (0%~100%)
def calculate_price(price, discount=0):
    discounted = price * (1 - discount)
    return discounted


print(f"기본 가격: {calculate_price(10000)}원")
print(f"할인 가격: {calculate_price(10000, 0.1)}원")
```

## 함수에서 여러값 반환하기

```python
# 최솟값과 최댓값을 동시 반환하는 함수
def min_max(numbers):
    minimum = min(numbers)
    maximum = max(numbers)
    return minimum, maximum


scores = [10,20,30,40,50,60,70,80]
low, high = min_max(scores)
print("최솟값:", low)  
print("최댓값:", high)  
```

## 문제

class_scores = [45,62,78,55,90,83,71,66,48,95]


성적 분석 함수 생성
아래와 같은 로그 출력
    - 평균: x.x 점(소수점 1번째까지만 출력)
    - 최고점: xx점
    - 최저점: xx점
    - 합격(60점 이상): x명

```python
class_scores = [45, 62, 78, 55, 90, 83, 71, 66, 48, 95]

def analyze_scores(scores):
    avg = sum(scores) / len(scores)
    high = max(scores)
    low = min(scores)

    count = 0
    for s in scores:
        if s >= 60:
            count += 1

    print(f"평균: {avg:.1f} 점")
    print("최고점:", high, "점")
    print("최저점:", low, "점")
    print("합격(60점 이상):", count, "명")


analyze_scores(class_scores)
```

```python
# 학생 데이터
grade_data = {
    "name": ["kim", "lee", "park", "cho", "choi", "jung"],
    "korean": [88      ,    72,          95      ,        65     ,    80      , 91],
    "math": [  75      ,    90,          65      ,        85     ,    77      , 81],
    "english": [  82      ,    40,          55      ,        25     ,    67      , 71],

}

grade_df= pd.DataFrame(grade_data)
grade_df
```

## 정렬하기

```python
print("----  국어 점수 낮은 순 -----")
print(grade_df.sort_values("korean"))
print()
print("----- 수학 점수 높은 순 -----")
print(grade_df.sort_values("math", ascending=False))
```

## 새로운 열 추가하기


```python
grade_df["average"] = \
    (grade_df["korean"] + \
        grade_df["math"] + \
            grade_df["english"]) / 3         # \ 다음 줄
grade_df["average"] = grade_df["average"].round(1)
grade_df
```

```python
grade_df["grade"] = grade_df["average"].apply(
    lambda x: "A" if x >= 80 else "B"
)

grade_df
```

## 그룹별 통계

```python
group_avg = grade_df.groupby("grade")["average"].mean()    # 위에 열을 가져와야 해서 average 대괄호
print("등급별 평균:")
print(group_avg)
```

```python
# 응용 - 회사 부서별 급여 파악

employee_data = {
    "name": ["kim", "lee", "park", "cho", "choi", "jung"],
    "department": ["개발", "마케팅", "개발", "인사", "마케팅", "인사"],
    "salary": [4200, 3500, 5000, 6000, 7000, 6500]

}
    
emp_df = pd.DataFrame(employee_data)
emp_df

# 부서별 평균 급여

print("---- 부서별 평균 급여 ----")
dept_avg = emp_df.groupby("department")["salary"].mean()
print(dept_avg.round(0))
print()
print("---- 부서별 인원 수 ----")
dept_count = emp_df.groupby("department")["name"].count()
print(dept_count)
```
